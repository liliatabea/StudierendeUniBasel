## Wie sieht die Entwicklung nach Fakultäten aus? 

Neuen DataFrame erstellen

```python
data_subjects = data.iloc[:,[0,10,13,16,19,22,25,28]]
data_subjects.head()
```

```python
data_subjects = data_subjects.melt('Jahr', var_name='Fach', value_name='Anzahl')
data_subjects.head()
```

```python
plt.figure(figsize = (15, 12))
plt.title('Anteil an Fächern, Veränderung über die Jahre')

sns.set_palette(['#1b9e77', '#d95f02', '#7570b3', '#e7298a', '#66a61e', '#e6ab02', '#a6761d'])

data_subjects['Fach'] = data_subjects['Fach'].str.replace('_total', '', case=False)

sns.lineplot(data = data_subjects, x='Jahr', y='Anzahl', hue='Fach')
sns.despine()
```

Grundsätzlich lässt sich sagen, dass über alle Fächer, Theologie ausgenommen, ein steiler Anstieg stattgefunden hat. Fokussieren wir auf die letzten 30 Jahre.

```python
subjects = subjects.melt('Jahr', var_name='Fach', value_name='Anzahl')

#Neue Spalte für Geschlech
subjects['Geschlecht'] = subjects['Fach'].str[-1:].astype(str)

subjects['Fach_alt'] = subjects['Fach']

#Spalte "Fach" korrigieren, damit nur Fach ohne Geschlecht
subjects['Fach'] = subjects['Fach'].str.replace('_m', '', case=False)
subjects['Fach'] = subjects['Fach'].str.replace('_f', '', case=False)

subjects = subjects[subjects['Jahr']>1990]

subjects.head()
#subjects['Geschlecht'].unique()
```

```python
subjects.shape
```

```python
f, ax = plt.subplots(figsize=(15, 12))
plt.title('Studierende nach Fakultät und Geschlecht')

sns.set_palette(['#59ACAC', '#C89E71'])

ax.tick_params(axis='x', rotation=85)

sns.boxplot(data = subjects, x='Fach', y='Anzahl', hue='Geschlecht')
sns.despine()
```

Auch in den letzten 30 Jahren gab es in jeder Fakultät bei der Anzahl Studentinnen eine viel grössere Spannweite als bei den Studenten.

```python
plt.figure(figsize = (15, 12))
plt.title('Studierende nach Fakultät und Geschlecht, zeitliche Entwicklung')

sns.set_palette(['#d95f02', '#fdcdac', '#a6761d', '#f1e2cc', '#1b9e77', '#b3e2cd', '#e7298a', '#f4cae4', '#66a61e', '#e6f5c9', '#e6ab02', '#fff2ae', '#7570b3', '#cbd5e8'])

sns.lineplot(data = subjects, x='Jahr', y='Anzahl', hue='Fach_alt')
sns.despine()
```

### Unterschiede in den drei Top-Fakultäten

```python
data = pd.read_csv('data/100191.csv', delimiter=';')
subjects_main = data.iloc[:,[0,14,15,17,18,26,27]]

subjects_main = subjects_main.melt('Jahr', var_name='Fach', value_name='Anzahl')

#Neue Spalte für Geschlech
subjects_main['Geschlecht'] = subjects_main['Fach'].str[-1:].astype(str)

subjects_main['Fach_alt'] = subjects_main['Fach']

#Spalte "Fach" korrigieren, damit nur Fach ohne Geschlecht
subjects_main['Fach'] = subjects_main['Fach'].str.replace('_m', '', case=False)
subjects_main['Fach'] = subjects_main['Fach'].str.replace('_f', '', case=False)

subjects_main = subjects_main[subjects_main['Jahr']>1990]

subjects_main.head()
```

```python
plt.figure(figsize = (15, 12))
plt.title('Studierende der drei meistgewählten Fakultäten')

sns.set_palette(['#1b9e77', '#b3e2cd', '#e7298a', '#f4cae4', '#7570b3', '#cbd5e8'])

sns.lineplot(data = subjects_main, x='Jahr', y='Anzahl', hue='Fach_alt')
sns.despine()
```
